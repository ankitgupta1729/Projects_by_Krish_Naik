### Part 1 - Data Preprocessing

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('BankCustomers.csv')
```

```python
dataset
```

```python
## exited column contains past information whether in past he has left the bank or not.
## hascrcard column denotes whether customer has credit card or not
X = dataset.iloc[:, 3:13]
y = dataset.iloc[:, 13]
```

```python
X
```

```python
## convert categorical feature into dummy variables
## if there are three classes in geography, then we assign 000,001,... to them for each combination and drop the first because
## we don't need the first class, we can understand whether first class is there or not by other two classes
## for gender, male and female are two classes..we drop the female class because we can get it by male class..if male=0
## then it means female is there otherwise not 

states=pd.get_dummies(dataset['Geography'],drop_first=True)
gender=pd.get_dummies(dataset['Gender'],drop_first=True)
```

```python
states
```

```python
gender
```

```python
#drop the columns as it is no longer required

X=X.drop(['Geography','Gender'],axis=1)
```

```python
X
```

```python
#concatenate the remaining dummies columns
X=pd.concat([X,states,gender],axis=1)
```

```python
X
```

```python
X
```

```python
y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### Part 2 - Now let's make the ANN!

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()
```

```python
## Adding the input layer and the first hidden layer
## "uniform" means weights are initialized 0 to 1.
## input neurons(features) are 11 and output neurons are 6 (just making ann in this way) and output function is relu
## to make output between 0 and 1
classifier.add(Dense(activation="relu", input_dim=11, units=6, kernel_initializer="uniform"))
```

```python
# Adding the second hidden layer
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))

# Adding the output layer
classifier.add(Dense(activation="sigmoid", units=1, kernel_initializer="uniform"))
```

```python
classifier.summary()
```

```python
## Compiling the ANN
## if we have more classes to classify then we can use categoricall_crossentropy 
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
## batch_size =10 means we are taking 10 different inputs
```

```python
# Fitting the ANN to the Training set
# epoch means combination of forward and backward propagation
classifier.fit(X_train, y_train, batch_size = 10, epochs = 50)
```

### Part 3 - Making the predictions and evaluating the model

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy=accuracy_score(y_test,y_pred)
```
