```python
## Usecase: Drug Vaccination between age 13-100 years
## people >65 yrs : observed some side effects
## people <65 yrs : no side effects
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import adam_v2
```

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_sample=[]
train_label=[]
```

```python
for i in range(1000):
    younger_ages=randint(13,64)
    train_sample.append(younger_ages)
    train_label.append(0)

    older_ages=randint(65,100)
    train_sample.append(older_ages)
    train_label.append(1)
```

```python
train_sample
```

```python
train_label
```

```python
train_sample=np.array(train_sample)
```

```python
train_label=np.array(train_label)
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
scaler_train_sample=scaler.fit_transform(train_sample.reshape(-1,1))
```

```python
scaler_train_sample
```

```python
model=Sequential([Dense(16,input_dim=1,activation='relu'),Dense(32,activation='relu'),Dense(2,activation='softmax')])
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(Adam(lr=0.001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
len(train_sample)
```

```python
len(train_label)
```

```python
model.fit(train_sample,train_label,batch_size=10,epochs=10)
```

```python
test_sample=[]
test_label=[]
```

```python
for i in range(500):
    younger_ages=randint(13,64)
    test_sample.append(younger_ages)
    test_label.append(0)

    older_ages=randint(65,100)
    test_sample.append(older_ages)
    test_label.append(1)
```

```python
test_sample=np.array(test_sample)
test_label=np.array(test_label)
```

```python
model.predict(test_sample,batch_size=10)
```

```python
test_sample_output=(model.predict(test_sample) > 0.5).astype("int32")
```

```python
test_output=[]
for i in range(len(test_sample_output)):
    test_output.append(test_sample_output[i][0])
test_output=np.array(test_output)
```

```python
from sklearn.metrics import confusion_matrix
predictedvalues=confusion_matrix(test_label,test_output)
```
